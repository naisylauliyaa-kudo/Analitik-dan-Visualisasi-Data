```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Pada tahap ini, kita akan melihat nilai ekstrem (outliers) pada kolom yang bertipe data kan numerik. Hal ini dilakukan agar mendapatkan insight yang akurat sehingga dapat memperkuat analisis dan model yang akan kita bangun nantinya.

## **Outliers**

Outliers adalah nilai data yang angkanya jauh lebih besar atau lebih kecil dibandingkan sebagian besar data lainnya. Sederhananya, outlier itu data yang “menyimpang” dari pola umum. Pada tahap ini kita harus mengetahui dulu outliers yang ada pada satu data terjadi dikarenakan hal apa, lalu kita putuskan apakah perlu dihapus, dibiarkan, atau ditransformasi agar tidak mempengaruhi hasil analisis.

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Berdasarkan hasil identifikasi menggunakan metode statistik dan visualisasi, ditemukan beberapa variabel dengan persentase outliers yang bervariasi, seperti online_only (21,90%), brand_id (16,78%), child_count (12,96%), reviews (11,33%), loves_count (11,19%), price_usd (9,94%), out_of_stock (7,36%), new (7,16%), dan limited_edition (7,02%), serta beberapa variabel lainnya dengan persentase di bawah 5%. Namun, penanganan tidak dilakukan secara seragam, melainkan disesuaikan dengan peran dan relevansi masing-masing variabel terhadap tujuan utama analisis, yaitu harga, rating, dan popularitas produk.  

Untuk variabel seperti online_only, child_count, out_of_stock, child_max_price, child_min_price, value_price_usd, sale_price_usd, limited_edition, serta new, tidak dilakukan treatment khusus terhadap outliers karena kolom-kolom tersebut akan dihapus pada tahap akhir pembersihan data. Variabel-variabel ini hanya merepresentasikan atribut atau status tambahan produk dan tidak termasuk dalam fokus utama analisis, yaitu harga, rating, dan popularitas. Oleh karena itu, keberadaan outliers di dalamnya tidak memerlukan penanganan lanjutan.

Khusus untuk brand_id, meskipun terdeteksi memiliki persentase outliers yang cukup tinggi, tidak dilakukan pemotongan (clipping) berdasarkan metode Interquartile Range (IQR). Hal ini karena brand_id merupakan variabel identitas (ID) yang bersifat nominal. Angka pada kolom tersebut hanya berfungsi sebagai label unik untuk setiap merek dan tidak memiliki makna matematis. Melakukan pemotongan justru berisiko mengubah identitas merek dan mengganggu proses mapping data pada tahap analisis berikutnya. Untuk menghindari kesalahan interpretasi sebagai variabel numerik, kolom ini dikonversi ke tipe data object.

Sementara itu, pada variabel utama seperti reviews, loves_count, rating, dan price_usd, nilai outliers sengaja dipertahankan (retained). Outliers pada variabel ini dikategorikan sebagai natural outliers karena mencerminkan kondisi pasar yang nyata. Dalam konteks e-commerce seperti Sephora, wajar apabila terdapat produk yang sangat populer atau viral sehingga memiliki jumlah ulasan dan interaksi jauh di atas rata-rata, serta produk dengan harga premium yang berada di luar rentang umum. Menghapus nilai-nilai tersebut justru dapat menghilangkan informasi penting terkait dinamika popularitas dan segmentasi harga produk di pasar.

Secara keseluruhan, pendekatan penanganan outliers dilakukan secara selektif dan berbasis relevansi, sehingga kualitas data tetap terjaga tanpa mengorbankan validitas informasi yang merepresentasikan kondisi pasar secara aktual.
